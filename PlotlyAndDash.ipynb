```python
from plotly.data import gapminder
import plotly.express as px
import pandas as pd
df=px.data.gapminder()
df
```

```python
px.scatter(df[df['year']==2007],x ="gdpPercap",y= "lifeExp",size= 'pop',color='continent' ,hover_name='country',log_x=True,title= 'my chart',trendline='ols')
```

```python
import plotly.io as pio
pio.templates
px.scatter(df,x ="gdpPercap",y= "lifeExp",size= 'pop',color='continent' ,template='plotly_dark',
           animation_frame='year',hover_name='country',log_x=True,title= 'my chart',trendline='ols')
```
