# 2)BFS Path for the graph

```python
graph={0:[1,3,8],
       1:[0,7],
       2:[7,5,3],
       3:[0,2,4],
       4:[3,8],
       5:[2,6],
       6:[5],
       7:[1,2],
       8:[0,4]}

def bfs(graph,start):
    visited = []
    qeue=[]
    visited.append(start)
    qeue.append(start)
    while len(qeue) !=0:
        node =qeue.pop(0)
        
        for edge in graph[node]:
            if edge not in visited:
                visited.append(edge)
                qeue.append(edge)
    return visited

bfs(graph,0)
#true output [0, 1, 3, 8, 7, 2, 4, 5, 6]
```

# 3) Dijkstra for the following weighted graph

```python
graph = {
                    0: {1:8, 2:9},
                    1: {0:8, 5:7, 3:1, 6:9},
                    2: {0:9, 5:6, 6:4},
                    3: {1:1, 7:4},
                    4: {7:7, 6:7},
                    5: {1:7, 2:6, 7:2},
                    6: {1:9, 2:4, 4:7, 7:6},
                    7: {3:4, 4:7, 5:2, 6:6}
                }
def dijsktras(graph,startNode,endNode):
    if(startNode in graph and endNode in graph ):
        shortestPath = {}
        shortestPathWeight ={}
        for keys in graph.keys():#initializa 
            shortestPathWeight[keys]=float("inf")
            shortestPath[keys] = []
        shortestPathWeight[startNode]=0
        myQueue = []#not need to priorty queue becouse we comber the weight drictly
      
        for items in graph[startNode].keys():
            myQueue.append(items)
            shortestPath[items].append(startNode)
            shortestPath[items].append(items)
            shortestPathWeight[items] = graph[startNode][items]
        while( len(myQueue)>0):
            currrentKey =myQueue.pop(0)
            for items,weights in graph[currrentKey].items():
                if(weights+shortestPathWeight[currrentKey]<shortestPathWeight[items]):#comper the weighe for the node from carant path and pirves
                    myQueue.append(items)
                    shortestPath[items]=[]
                    shortestPathWeight[items] = weights+shortestPathWeight[currrentKey]
                    shortestPath[items].extend(shortestPath[currrentKey])
                    #for x in shortestPath[currrentKey]:# .extend put the shortest path for cruntKay after discver it was the best 
                     #   shortestPath[items].append(x)
                    shortestPath[items].append(items)


        if((shortestPath[endNode] != [])or(endNode==startNode)):
            print("with a weight of "+str(shortestPathWeight[endNode])+" shortest path for "+str(endNode)+ " from "+str(startNode)+" is: ")
            print(shortestPath[endNode])

        else:
            print("could not find a path to "+str( endNode )+" from "+str(startNode))
    else:
        print("cannot locate start or end node")
dijsktras(graph, 0, 6)
```
