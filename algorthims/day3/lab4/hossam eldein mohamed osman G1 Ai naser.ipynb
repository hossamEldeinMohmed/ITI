```python
import networkx as nx
```

# 1- Draw Graph

```python
d={0:[1,3,8],
   1:[0,7],
   2:[7,5,3],
   3:[0,2,4],
   4:[3,8],
   5:[2,6],
   6:[5],
   7:[1,2],
   8:[0,4]}
g=nx.Graph(d)
nx.draw(g,with_labels = True)
```

# 2) DFS path for the Graph

```python
dfs = list(nx.dfs_tree(g, source=0))

print(dfs)
```


# 3)BFS Path for the graph

```python
dfs = list(nx.bfs_tree(g, source=0))

print(dfs)
```

# 4) Dijkstra for the following weighted graph

```python
import matplotlib.pyplot as plt
```

```python

nodes = {
    0: [1, 2],
    1: [0, 5, 3, 6],
    2: [0, 5, 6],
    3: [1, 7],
    4: [7, 6],
    5: [1, 2, 7],
    6: [1, 4, 7],
    7: [3, 4, 5, 6]
}
G = nx.Graph(nodes)

G.add_edge(0,1,weight=8)
G.add_edge(0,2,weight=9)
G.add_edge(1,3,weight=1)
G.add_edge(1,5,weight=7)
G.add_edge(1,6,weight=9)
G.add_edge(2,5,weight=6)
G.add_edge(2,6,weight=4)
G.add_edge(3,7,weight=4)
G.add_edge(4,7,weight=7)
G.add_edge(4,6,weight=7)
G.add_edge(5,7,weight=2)
G.add_edge(6,7,weight=6)
pos = nx.spring_layout(G)
nx.draw(G,pos,with_labels=True)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)

plt.show()
```

```python
distance, path = nx.single_source_dijkstra(G, source=0, target=7)
print(path)
print(distance)
```
